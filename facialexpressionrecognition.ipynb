```python
# data preparation

train_dir = '/kaggle/input/facial-recognition-dataset/Training/Training'
test_dir = '/kaggle/input/facial-recognition-dataset/Testing/Testing'

train_angry_dir = '/kaggle/input/facial-recognition-dataset/Training/Training/Angry'
train_fear_dir = '/kaggle/input/facial-recognition-dataset/Training/Training/Fear'
train_happy_dir = '/kaggle/input/facial-recognition-dataset/Training/Training/Happy'
train_neutral_dir = '/kaggle/input/facial-recognition-dataset/Training/Training/Neutral'
train_sad_dir = '/kaggle/input/facial-recognition-dataset/Training/Training/Sad'
train_surprise_dir = '/kaggle/input/facial-recognition-dataset/Training/Training/Suprise'

test_angry_dir = '/kaggle/input/facial-recognition-dataset/Testing/Testing/Angry'
test_fear_dir = '/kaggle/input/facial-recognition-dataset/Testing/Testing/Fear'
test_happy_dir = '/kaggle/input/facial-recognition-dataset/Testing/Testing/Happy'
test_neural_dir = '/kaggle/input/facial-recognition-dataset/Testing/Testing/Neutral'
test_sad_dir = '/kaggle/input/facial-recognition-dataset/Testing/Testing/Sad'
test_surprise_dir = '/kaggle/input/facial-recognition-dataset/Testing/Testing/Suprise'

dir_list = [train_angry_dir, train_fear_dir, train_happy_dir,
           train_neutral_dir, train_sad_dir, train_surprise_dir,
           test_angry_dir, test_fear_dir, test_happy_dir,
           test_neural_dir, test_sad_dir, test_surprise_dir]

for d in dir_list:
    print(d, len(os.listdir(d)))
```

```python
# Data augmentation to generate additional training data
train_datagen = ImageDataGenerator(
    rescale=1.0/255.0,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

test_datagen = ImageDataGenerator(rescale=1.0/255.0)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(150,150),
    color_mode='grayscale',
    batch_size=32,
    class_mode='categorical'
)
validation_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(150,150),
    color_mode='grayscale',
    batch_size=32,
    class_mode='categorical'
)
train_datagen = ImageDataGenerator(rescale=1.0/255.0,
                                  rotation_range=40,
                                  width_shift_range=0.2,
                                  height_shift_range=0.2,
                                  zoom_range=0.2,
                                  horizontal_flip=True,
                                  fill_mode='nearest')

train_generator = train_datagen.flow_from_directory(train_dir,
                                                   target_size=(150,150),
                                                   batch_size=64,
                                                   class_mode='categorical')

validation_datagen = ImageDataGenerator(rescale=1.0/255.0)

validation_generator = validation_datagen.flow_from_directory(test_dir,
                                                             target_size=(150,150),
                                                             batch_size=62,
                                                             class_mode='categorical')
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.optimizers import RMSprop
```

```python
early_stopping_cb = tf.keras.callbacks.EarlyStopping(patience=5,
                                                     restore_best_weights=True)
```

```python

model = tf.keras.models.Sequential([
    Conv2D(16, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    MaxPooling2D(2, 2),
    Conv2D(32, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Flatten(),
    Dense(512, activation='relu'),
    Dense(6, activation='softmax')
])

model.compile(optimizer=RMSprop(learning_rate=0.001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
#training 
history = model.fit(
    train_generator,
    epochs=10,
    verbose=1,
    validation_data=validation_generator,
    callbacks=early_stopping_cb
)
```

```python
#Evaluate the model on the testing dataset
scores = model.evaluate_generator(validation_generator, validation_generator.n // validation_generator.batch_size)
print('Test loss:', scores[0])
print('Test accuracy:', scores[1])
```

```python
#Save the trained model
model.save('facial_expression_model.h5')
```
